```python
import pandas as pd
violations_path = '/NH_Violations__2022-2023_.csv'
license_path = '/Licence_Type__2019-2023_.csv'

violations = pd.read_csv(violations_path)
license_df = pd.read_csv(license_path)

print("Violations rows:", len(violations))
print(violations.head(10).to_string(index=False))
print("\nLicence data rows:", len(license_df))
print(license_df.to_string(index=False))
```

```python
# Compute % change and shares for violations
violations['Accidents_%_change'] = ((violations['Accidents_2023'] - violations['Accidents_2022']) / violations['Accidents_2022'] * 100).round(1)
violations['Fatalities_%_change'] = ((violations['Fatalities_2023'] - violations['Fatalities_2022']) / violations['Fatalities_2022'] * 100).round(1)

total_acc_2023 = int(violations.loc[violations['Category']=='All India', 'Accidents_2023'].values[0])
violations['Accidents_2023_share_pct'] = (violations['Accidents_2023'] / total_acc_2023 * 100).round(1)

# Licence share (2023)
if '2023' in license_df.columns:
    total_lic_2023 = int(license_df.loc[license_df['Type_of_Licence']=='Total', '2023'].values[0])
    license_df['share_2023_pct'] = (license_df['2023'] / total_lic_2023 * 100).round(1)
else:
    license_df['share_2023_pct'] = None

violations.head(8)
```

```python
import plotly.express as px

cats = violations[violations['Category'] != 'All India'].sort_values('Accidents_2023', ascending=False)
fig = px.bar(
    cats,
    x='Category',
    y='Accidents_2023',
    hover_data=['Accidents_2022','Fatalities_2022','Fatalities_2023','Accidents_%_change','Accidents_2023_share_pct'],
    title='Accidents by Category (2023)'
)
fig.update_layout(xaxis_tickangle=-45, margin=dict(t=60, b=160))
fig.show()
```

```python
line_df = cats.melt(id_vars=['Category'], value_vars=['Accidents_2022','Accidents_2023'], var_name='Year', value_name='Accidents')
line_df['Year'] = line_df['Year'].str.replace('Accidents_','')
fig2 = px.line(line_df, x='Category', y='Accidents', color='Year', markers=True, title='Accidents: 2022 vs 2023')
fig2.update_layout(xaxis_tickangle=-45)
fig2.show()
```

```python
years = ['2019','2020','2021','2022','2023']
lic_plot = license_df[license_df['Type_of_Licence'] != 'Total'].copy()
plot_df = lic_plot.melt(id_vars=['Type_of_Licence'], value_vars=years, var_name='Year', value_name='Accidents')
fig3 = px.line(plot_df, x='Year', y='Accidents', color='Type_of_Licence', markers=True, title='Licence-type accidents (2019-2023)')
fig3.show()
```
